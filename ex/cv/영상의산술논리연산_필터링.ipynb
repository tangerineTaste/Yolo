```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
src1 = cv2.imread('./img/lenna256.bmp', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('./img/square.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
dst1 = cv2.add(src1, src2)
dst2 = cv2.addWeighted(src1, 0.5, src2, 0.5, 0.0)
dst3 = cv2.subtract(src1, src2)
dst4 = cv2.absdiff(src1, src2)
```

```python
plt.subplot(231), plt.axis('off'), plt.imshow(src1,'gray'), plt.title('src1')
plt.subplot(232), plt.axis('off'), plt.imshow(src2,'gray'), plt.title('src2')
plt.subplot(233), plt.axis('off'), plt.imshow(dst1,'gray'), plt.title('dst1')
plt.subplot(234), plt.axis('off'), plt.imshow(dst2,'gray'), plt.title('dst2')
plt.subplot(235), plt.axis('off'), plt.imshow(dst3,'gray'), plt.title('dst3')
plt.subplot(236), plt.axis('off'), plt.imshow(dst4,'gray'), plt.title('dst4')
plt.show()
```

```python
wing = cv2.imread('./img/wing_wall.jpg', cv2.IMREAD_GRAYSCALE)
yate = cv2.imread('./img/yate.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img = cv2.addWeighted(wing,0.7, yate,0.5,0)
```

```python
gmin = float(np.min(img))
gmax = float(np.max(img))
print(gmin, gmax)
```

```python
dst = ((img - gmin) * 255. / (gmax - gmin)).astype('uint8')
```

```python
cv2.imshow('dst',dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
lion = cv2.imread('./img/lion_face.jpg', cv2.COLOR_BGR2RGB)
man = cv2.imread('./img/man_face.jpg', cv2.COLOR_BGR2RGB)
dst_face = cv2.addWeighted(lion,1.0, man,0.0,0) 
```

```python
def onChange(x):
    try:
        x = cv2.getTrackbarPos('t','dst') 
        alpha = x/100
    except:
        x =0 
    dst_face = cv2.addWeighted(lion,1-alpha, man,alpha,0) 
    cv2.imshow('dst',dst_face)
```

```python
cv2.imshow('dst',dst_face)                                

cv2.createTrackbar('t', 'dst', 0, 100, onChange)

while True:
    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()        
```

```python
robot1 = cv2.imread('./img/robot_arm1.jpg', cv2.COLOR_BGR2RGB)
robot2 = cv2.imread('./img/robot_arm2.jpg', cv2.COLOR_BGR2RGB)
```

```python
dif = cv2.subtract(robot2,robot1)
```

```python
dif = dif.astype(np.uint8)
```

```python
red = np.full(320*480*3,0).reshape(320,480,3)
```

```python
red[:,:] = [0,0,255]
```

```python
red = red.astype(np.uint8)
```

```python
result = cv2.multiply(red, dif)
```

```python
result = cv2.add(robot1,result)
```

```python
cv2.imshow('t',result)
cv2.waitKey()
cv2.destroyAllWindows()
```
